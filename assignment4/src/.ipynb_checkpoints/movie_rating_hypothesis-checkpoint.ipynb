##### Read and cleanup the movie data

```python
def get_movie_data():
    
    unames = ['user_id','gender','age','occupation','zip']
    users = pd.read_table(os.path.join('../data','users.dat'), 
                          sep='::', header=None, names=unames, encoding='latin-1')
    
    rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
    ratings = pd.read_table(os.path.join('../data', 'ratings.dat'), 
                            sep='::', header=None, names=rnames, encoding='latin-1')
    
    mnames = ['movie_id', 'title','genres']
    movies = pd.read_table(os.path.join('../data', 'movies.dat'), 
                           sep='::', header=None, names=mnames, encoding='latin-1')

    return users, ratings, movies

users, ratings, movies = get_movie_data()

tmp = movies.title.str.extract('(.*) \(([0-9]+)\)')
tmp.apply(lambda x:x[0] if len(x) > 0 else None)
tmp.apply(lambda x: x[0][:40] if len(x) > 0 else None)

movies['year'] = tmp[1]
movies['short_title'] = tmp[0]

print(movies.head())
```

### 1. Prove or disprove the following Null Hypothesis:

#### H0: Different age segments of users rate the different genres of movies similarly

- Age segments are (0-20), (21-30), (31-50), (50 and above),
- Genres - Drama, Comedy, Action, Romance, Adventure
- Rating 4 or above is a positive, else negative (used for binary votes)

#### Solution

# I Implemented the chi2 test of independence in this assignment

```python
print(users.head())
```

```python
print(ratings.head())
```

```python
print(movies.head())
```

```python
ratings_users = pd.merge(users,ratings, left_on='user_id', right_on='user_id')
ratings_users_movies = pd.merge(movies,ratings_users, left_on='movie_id' , right_on = 'movie_id')

ratings_users_movies
```

- codification of ratings label
- Rating 4 or above is a positive, else negative (used for binary votes)
- ratings of 4 and 5 will be '1'(positive) and lower ratings are '0' (negitive)

```python
cat = []
for i in range (0, 1):
    for value in ratings_users_movies["rating"]:
        if value >= 4 :
            cat.append("1")
        elif value < 4:
            cat.append("0")

```

```python
ratings_users_movies["ratings_label"] = cat
```

```python
ratings_users_movies.head(10)
```

- **now codification of age into 4 groups**

```python
group = []
for i in range (0, 1):
    for value in ratings_users_movies["age"]:
        if value >= 0 and value <= 20:
            group.append("1")
        elif value >= 21 and value <= 30:
            group.append("2")
        elif value >= 31 and value <= 50 :
            group.append("3")
        elif value > 50 :
            group.append("4")

        
    
#df_n["cat"] = cat   

ratings_users_movies["age_label"] = group
```

- Age segments are (0-20), (21-30), (31-50), (50 and above),
- (0-20) is 1
- (21 - 30) is 2
- (31 - 50) is 3
- (51 and above) is 4


```python
ratings_users_movies
```

- Extracting required genres from the dataset
- **Genres - Drama, Comedy, Action, Romance, Adventure**

```python
genres = ratings_users_movies.loc[ratings_users_movies['genres'].str.contains('Drama'and'Action'and'Adventure'and'Comedy'and'Romance')]

#genres = ratings_users_movies.loc[ratings_users_movies.genres == (['Drama'|'Action'|'Adventure'|'Comedy'|'Romance'])]

#genres = ratings_users_movies.loc[ratings_users_movies['genres'].isin(['Drama', 'Action', 'Adventure', 'Comedy', 'Romance'])]


```

```python
genres
```

```python
final = genres[genres['ratings_label']=='1']
# extraxting the dataframe with ratings positive ratings only
```

- pd.crosstab method is used to compute a simple cross-tabulation of two (or more) factors. 
- By default, it computes a frequency table of the factors unless an array of values and an aggregation function are passed

```python
table = pd.crosstab(final.genres,final.age_label, margins=True)
```

```python
table
# all the positive ratings are only considered
# for all the genres four age segments of user are extracted
```

```python
table.shape
```

```python
newdf = table.drop("All", axis='columns')
newdf = newdf.drop('All', axis='index')

# from the table dropping the unwanted rows and columns
```

```python
newdf.shape
```

```python


newdf.columns = ["age(0-20)","age(21-30)","age(31-50)","age(51 and above)"]

# reassigning the column names for the table below
```

```python
newdf
```

```python
stats.chi2_contingency(observed= newdf)
```

- 'p' value is 0
- chi2 statistic value = 2824.9
- Degrees of Freedom = 174
- expected array - same shape as the observed data table based on marginal sums of the table

```python
crit = stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence
                      df = 174)
```

```python
print('Critical value is:\n', crit)
# critical value is given below
```

# we reject the null hypothesis as the p-value obtained is '0'
